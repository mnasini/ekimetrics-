```python
import pandas as pd
```

```python
contacts=pd.read_excel("data/Contacts.xlsx")
```

```python
don_2018=pd.read_excel("data/Dons_2018.xlsx")
don_2017=pd.read_excel("data/Dons_2017.xlsx")
don_2016=pd.read_excel("data/Dons_2016.xlsx")
don_2015=pd.read_excel("data/Dons_2015.xlsx")
don_2015=pd.read_excel("data/Dons_2014.xlsx")
```

```python
donors=pd.concat([don_2018,don_2017,don_2016,don_2015])
```

```python
donations=donors.merge(contacts,how="inner",left_on="contact ID",right_on="ID_CTC")
```
